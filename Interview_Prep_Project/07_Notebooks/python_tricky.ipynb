## Q1: `is` vs `==` (Identity vs Equality)

```python
# Predict: What does this print?
a = 256
b = 256
print(f"a == b: {a == b}")
print(f"a is b: {a is b}")
```

```python
# Now try with 257:
a = 257
b = 257
print(f"a == b: {a == b}")
print(f"a is b: {a is b}")

# WHY? Python caches integers from -5 to 256!
```

## Q2: Mutable Default Arguments

```python
# PREDICT: What gets printed each time?

def add_item_bad(item, items=[]):
    items.append(item)
    return items

print(add_item_bad('a'))  # ?
print(add_item_bad('b'))  # ?
print(add_item_bad('c'))  # ?
```

```python
# THE FIX:

def add_item_good(item, items=None):
    if items is None:
        items = []
    items.append(item)
    return items

print(add_item_good('a'))  # ['a']
print(add_item_good('b'))  # ['b']
print(add_item_good('c'))  # ['c']

# RULE: Never use [] or {} as default parameter!
```

## Q3: List Copy vs Reference

```python
# PREDICT: What is a after this?

a = [1, 2, 3]
b = a
b.append(4)

print(f"a = {a}")
print(f"b = {b}")
```

```python
# How to COPY:

a = [1, 2, 3]
b = a.copy()     # Method 1
b = a[:]         # Method 2
b = list(a)      # Method 3

b.append(4)
print(f"a = {a}")  # Unchanged!
print(f"b = {b}")
```

```python
# But shallow copy fails for nested!
import copy

nested = [[1, 2], [3, 4]]
shallow = nested.copy()
deep = copy.deepcopy(nested)

nested[0][0] = 999

print(f"Original: {nested}")
print(f"Shallow:  {shallow}")  # Also changed!
print(f"Deep:     {deep}")      # Unchanged
```

## Q4: Closures & Late Binding

```python
# PREDICT: What does this print?

funcs = []
for i in range(3):
    funcs.append(lambda: i)

print([f() for f in funcs])  # ?

# WHY? Lambda captures the VARIABLE, not the VALUE
```

```python
# THE FIX:

funcs = []
for i in range(3):
    funcs.append(lambda x=i: x)  # Capture value as default

print([f() for f in funcs])  # [0, 1, 2]
```

## Q5: Variable Scoping (LEGB)

```python
# LEGB = Local, Enclosing, Global, Built-in

x = "global"

def outer():
    x = "enclosing"
    
    def inner():
        x = "local"
        print(f"inner: {x}")
    
    inner()
    print(f"outer: {x}")

outer()
print(f"global: {x}")
```

```python
# WHAT'S WRONG HERE? (UnboundLocalError)

x = 10

def broken():
    print(x)  # Python sees x = 5 below, assumes x is local
    x = 5     # But we try to print BEFORE assignment!

# broken()  # Uncomment to see error

# FIX: Use 'global x' or restructure code
```

## Q6: Truthiness

```python
# What's Falsy in Python?

falsy = [None, False, 0, 0.0, "", [], {}, set(), ()]

print("Falsy values:")
for val in falsy:
    print(f"  bool({repr(val):10}) = {bool(val)}")
```

```python
# Common mistake: 0 is falsy!

count = 0

if count:
    print("Count has a value")
else:
    print("Oops! count=0 is falsy")

# FIX: Explicit check
if count is not None:
    print(f"Count is {count}")
```

## Q7: Dict Key Gotchas

```python
# PREDICT: How many keys?

d = {
    1: "int",
    1.0: "float",
}

print(d)
print(f"Keys: {list(d.keys())}")

# WHY? 1 == 1.0 is True, so they're the same key!
```

```python
# True == 1 and False == 0

d = {
    True: "true",
    1: "one",
}

print(d)  # {True: 'one'}
```

## Q8: Generator vs List

```python
import sys

# List comprehension
list_comp = [x**2 for x in range(1000)]

# Generator expression
gen_exp = (x**2 for x in range(1000))

print(f"List type: {type(list_comp)}")
print(f"Gen type:  {type(gen_exp)}")
print(f"\nList memory: {sys.getsizeof(list_comp)} bytes")
print(f"Gen memory:  {sys.getsizeof(gen_exp)} bytes")

# Use generators for large data!
```

## Summary Cheat Sheet

| Gotcha | Rule |
|--------|------|
| `is` vs `==` | Use `is` only for None/True/False |
| Mutable defaults | Never use `[]` or `{}` as default |
| List copy | Use `.copy()` or `deepcopy()` |
| Closures | Capture value with `lambda x=i: x` |
| Truthiness | 0, "", [], {} are all Falsy |
| Dict keys | 1 == 1.0, True == 1 |
| Generators | Use `()` for memory efficiency |
